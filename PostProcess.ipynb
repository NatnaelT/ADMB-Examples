```
#!/usr/bin/env python2
# -*- coding: utf-8 -*-
"""
Created on Mon Jul 30 16:21:05 2018

@author: natnael.hamda
"""

import pandas as pd

import numpy as np


import os  


import glob


#import all the mean, variance & correlation 

#from pomegranate import *

#from calc_ll_deri import *

import pandas as pd

import numpy as np

from numpy import genfromtxt


import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np


from sklearn.mixture import GMM



#data_model_df = pd.read_csv("/Users/natnael.hamda/GoogleDrive/git/myProjects/behaviouralMap/GGS/batch_of_tracks/summary_stat_simu/very_final_simu_behavious_20180730/out_put_4-GMM/final_class_input_simu.csv")


data_model_df_2011 = pd.read_csv("Final_GMM_out_2011.csv")

data_model_df_2012 = pd.read_csv("Final_GMM_out_2012.csv")




data_model_gmm_input_2011 = data_model_df_2011.iloc[:,:9].values


data_model_gmm_input_2012 = data_model_df_2012.iloc[:,:9].values


data_model_gmm_input_all = np.vstack((data_model_gmm_input_2011, data_model_gmm_input_2012))




#data_model_gmm_input = pd.DataFrame(data_model_df_2011['var1_mu_V','Var2_mu_a' ]) #.iloc[:,1:])


n_components = np.arange(1, 20)
models = [GMM(n, covariance_type='diag', random_state=0).fit(data_model_gmm_input_2011)
          for n in n_components]

plt.plot(n_components, [m.bic(data_model_gmm_input_2011) for m in models], label='BIC:2011')
plt.plot(n_components, [m.aic(data_model_gmm_input_2011) for m in models], label='AIC:2011')
plt.legend(loc='best')
plt.xlabel('n_components');

plt.show()

n_components = np.arange(1, 20)
models = [GMM(n, covariance_type='diag', random_state=0).fit(data_model_gmm_input_2012)
          for n in n_components]

plt.plot(n_components, [m.bic(data_model_gmm_input_2012) for m in models], label='BIC:2012')
plt.plot(n_components, [m.aic(data_model_gmm_input_2012) for m in models], label='AIC:2012')
plt.legend(loc='best')
plt.xlabel('n_components');


plt.show()

n_components = np.arange(1, 20)
models = [GMM(n, covariance_type='diag', random_state=0).fit(data_model_gmm_input_all)
          for n in n_components]

plt.plot(n_components, [m.bic(data_model_gmm_input_all) for m in models], label='BIC:all')
plt.plot(n_components, [m.aic(data_model_gmm_input_all) for m in models], label='AIC:all')
plt.legend(loc='best')
plt.xlabel('n_components');

plt.show()
# 8 total bevaiours


from sklearn.manifold import TSNE

plt.figure()
# perform dimensionality reduction
tsne = TSNE()
transformed = tsne.fit_transform(data_model_gmm_input_all)




# visualize the clouds in 2-D
plt.scatter(transformed[:,0], transformed[:,1])
plt.show()
```
